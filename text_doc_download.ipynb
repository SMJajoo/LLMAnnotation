```python
import os
import shutil
import random

def merge_and_rename_folders(folder1, folder2, output_folder):
    # Create output folder if it doesn't exist
    os.makedirs(output_folder, exist_ok=True)

    # Collect all file paths from both folders
    files = []
    for folder in [folder1, folder2]:
        if os.path.exists(folder):
            files.extend([os.path.join(folder, f) for f in os.listdir(folder) if os.path.isfile(os.path.join(folder, f))])

    # Shuffle files randomly
    random.shuffle(files)

    # Copy and rename files in the new folder
    for index, file_path in enumerate(files, start=1):
        ext = os.path.splitext(file_path)[1]  # Get original extension
        new_filename = os.path.join(output_folder, f"{index}{ext}")
        shutil.copy(file_path, new_filename)

    print(f"All files have been merged into '{output_folder}' and renamed sequentially.")

# Example usage
merge_and_rename_folders(r"F:\DataScience\LLMAnnotation\film_wiki_texts", r"F:\DataScience\LLMAnnotation\finance_wiki_texts", "merged_folder")
```
